## 3\. Strings

Aplique três métodos distintos na *string* abaixo, você pode conferir alguns métodos neste [link](https://www.w3schools.com/python/python_ref_string.asp):

```
cancao = 'Roda mundo, roda gigante, roda moinho, roda pião.'

# 1 Cria nova variável que armazena a frase em letras maiúsculas
cancao_upper = cancao.upper()

# 2 Conta a quantidade de vezes que a palavra 'RODA' aparece na frase
qte_roda_na_cancao = cancao_upper.count('RODA')

# Imprime a informação
print(f"A palavra 'roda' aparece {qte_roda_na_cancao} vezes em '{cancao}'")

# 3 Substitui a primeira palavra 'roda' por 'gira'
nova_cancao = cancao_upper.replace('RODA', 'GIRA', 1)

# 4 Retorna a frase para a formatação original
nova_cancao = nova_cancao.capitalize()
print(nova_cancao)

```

Extraia da string abaixo o valor da taxa **selic** na variável `selic` e o valor do **ano** na variavel `ano`. Imprima os valores na tela.

```
noticia = 'Selic vai a 2,75% e supera expectativas; é a primeira alta em 6 anos.'

# Encontra a posição da taxa selic
posicao_selic_inicial = noticia.find('2,75%')
posicao_selic_final = noticia.find('%')

# Atribui substring à variável selic
selic = noticia[posicao_selic_inicial:posicao_selic_final + 1]
print(selic)


# Encontra a posição do ano
posicao_ano = noticia.find('6')
ano = noticia[posicao_ano]
print(ano)
```
